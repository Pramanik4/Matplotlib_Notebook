```
import matplotlib.pyplot as mp
C = ["B.COM","BSC IT","BSC CS","BBI"]
g = [20,45,50,30]
b = [25,50,60,20]
mp.bar(C,g,0.4,label="Girl")
mp.bar(C,b,0.4,bottom=g,label="Boy")
mp.title("Details")
mp.xlabel("Course")
mp.ylabel("Number")
mp.legend()
mp.show()
```

```
import matplotlib.pyplot as mp
import numpy as np
Medal = ["India","USA","China"]
gold = [1,39,38]
silver = [2,41,32]
bronze = [4,33,18]
b = list(np.add(gold,silver))
mp.bar(Medal,gold,0.4,label="Gold")
mp.bar(Medal,silver,0.4,bottom=gold,label="Silver")
mp.bar(Medal,bronze,0.4,bottom=b,label="Brronze")
mp.title("Tokyo Olympics 2021 Medal")
mp.xlabel("No of Medals")
mp.ylabel("Countries")
mp.legend()
mp.show()
```

```
# Project Stock Market
import matplotlib.pyplot as mp
sub = ["USA","UK","China","India","Brazil","Indonesia"]
marks = [55,33,13,3,2,1]
col = ["purple","indigo","blue","green","yellow","red"]
mp.title("Stock Market Paticipation",size=20,color="Orange",pad=20)
mp.bar(sub,marks,color=col)
mp.xlabel("Country",size=12,color="orange",labelpad=18)
mp.ylabel("Percentage",size=12,color="orange",labelpad=18)
mp.show()
```
